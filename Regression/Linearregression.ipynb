```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# Load Indian roads dataset
df = pd.read_csv('/content/indian_roads_dataset.csv')

# See first 5 rows
print(df.head())

# See shape — rows and columns
print("\nShape:", df.shape)

# See all columns
print("\nColumns:", df.columns.tolist())
```

```python
# Convert text columns to numbers
df['road_type'] = df['road_type'].map({
    'highway': 2,
    'urban': 1,
    'rural': 0
})

df['weather'] = df['weather'].map({
    'fog': 2,
    'rain': 1,
    'clear': 0
})

df['visibility'] = df['visibility'].map({
    'low': 2,
    'medium': 1,
    'high': 0
})

df['traffic_density'] = df['traffic_density'].map({
    'high': 2,
    'medium': 1,
    'low': 0
})

# Drop missing values
df = df.dropna(subset=['road_type', 'weather',
                        'visibility', 'traffic_density',
                        'risk_score'])

print("Data cleaned successfully")
print("Remaining rows:", len(df))
```

```python
# Input — only hour
X_simple = df[['hour']]

# Output — risk score
y = df['risk_score']

# Split — 80% train, 20% test
X_train, X_test, y_train, y_test = train_test_split(
    X_simple, y, test_size=0.2, random_state=42
)

# Train model
simple_model = LinearRegression()
simple_model.fit(X_train, y_train)

# Predict
simple_pred = simple_model.predict(X_test)

# Accuracy
simple_r2 = r2_score(y_test, simple_pred)

print("Simple Linear Regression")
print("R2 Score:", round(simple_r2, 3))
```
