```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder

# Load dataset
df = pd.read_csv('/content/indian_roads_dataset.csv')
print("Dataset loaded successfully")
print("Shape:", df.shape)
```

```python
# See categorical columns
print("Categorical columns:")
print(df.select_dtypes(include='object').columns.tolist())

# See unique values in each
print("\nWeather values:", df['weather'].unique())
print("Road type values:", df['road_type'].unique())
print("Visibility values:", df['visibility'].unique())
print("Traffic density values:", df['traffic_density'].unique())
```

```python
# Create dummy variables using pandas
df_dummies = pd.get_dummies(df, columns=['weather', 'road_type',
                                          'visibility', 'traffic_density'])

print("Before encoding columns:", 24)
print("After encoding columns:", len(df_dummies.columns))
print("\nNew columns:")
print(df_dummies.columns.tolist())
```

```python
# Avoid dummy variable trap using drop_first=True
df_no_trap = pd.get_dummies(df, columns=['weather', 'road_type',
                                          'visibility', 'traffic_density'],
                                          drop_first=True)

print("Without drop_first columns:", len(df_dummies.columns))
print("With drop_first columns:", len(df_no_trap.columns))
print("\nDummy variable trap avoided successfully")
```

```python
# Select categorical columns
cat_columns = ['weather', 'road_type', 'visibility', 'traffic_density']

# Create encoder
encoder = OneHotEncoder(sparse_output=False, drop='first')

# Fit and transform
encoded = encoder.fit_transform(df[cat_columns])

# See encoded output
print("Encoded shape:", encoded.shape)
print("\nCategories:")
for col, cats in zip(cat_columns, encoder.categories_):
    print(f"  {col}: {cats}")

# Convert to dataframe
encoded_df = pd.DataFrame(encoded,
             columns=encoder.get_feature_names_out(cat_columns))
print("\nEncoded dataframe:")
print(encoded_df.head())
```
