```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import math
%matplotlib inline

df = pd.read_csv('/content/indian_roads_dataset.csv')
df.head()
```

```python
# Fatal = 1, Not Fatal = 0
df['is_fatal'] = df['accident_severity'].apply(
    lambda x: 1 if x == 'fatal' else 0
)

print(df['is_fatal'].value_counts())
```

```python
# Input — hour only (same as sir used age only)
X = df[['hour']]

# Output — fatal or not
y = df['is_fatal']

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size=0.8
)

print("X_test:")
print(X_test.head())
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
print("Model trained successfully")
```

```python
y_predicted = model.predict(X_test)
print("Predictions:")
print(y_predicted[:6])
```

```python
print("Predict Probability:")
print(model.predict_proba(X_test)[:6])
```

```python
score = model.score(X_test, y_test)
print("Model Score:", score)
```

```python
print("Coefficient (m):", model.coef_)
print("Intercept (b):", model.intercept_)
```

```python
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def prediction_function(hour):
    m = model.coef_[0][0]
    b = model.intercept_[0]
    z = m * hour + b
    y = sigmoid(z)
    return y

# Test — is accident at 2am fatal?
hour = 2
result = prediction_function(hour)
print("Hour:", hour)
print("Probability:", round(result, 3))

if result >= 0.5:
    print("Prediction: FATAL accident likely")
else:
    print("Prediction: NOT FATAL accident likely")
```
