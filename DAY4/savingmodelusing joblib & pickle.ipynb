```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# Load Indian roads dataset
df = pd.read_csv('/content/indian_roads_dataset.csv')

# See first 5 rows
print(df.head())

# See shape — rows and columns
print("\nShape:", df.shape)

# See all columns
print("\nColumns:", df.columns.tolist())
```

```python
# Convert text columns to numbers
df['road_type'] = df['road_type'].map({
    'highway': 2,
    'urban': 1,
    'rural': 0
})

df['weather'] = df['weather'].map({
    'fog': 2,
    'rain': 1,
    'clear': 0
})

df['visibility'] = df['visibility'].map({
    'low': 2,
    'medium': 1,
    'high': 0
})

df['traffic_density'] = df['traffic_density'].map({
    'high': 2,
    'medium': 1,
    'low': 0
})

# Drop missing values
df = df.dropna(subset=['road_type', 'weather',
                        'visibility', 'traffic_density',
                        'risk_score'])

print("Data cleaned successfully")
print("Remaining rows:", len(df))
```

```python
# Multiple inputs
X_multi = df[['hour', 'weather', 'road_type',
              'visibility', 'traffic_density',
              'is_weekend', 'is_peak_hour']]

# Output
y = df['risk_score']

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_multi, y, test_size=0.2, random_state=42
)

# Train model
multi_model = LinearRegression()
multi_model.fit(X_train, y_train)

# Predict
multi_pred = multi_model.predict(X_test)

# Accuracy
multi_r2 = r2_score(y_test, multi_pred)

print("Multiple Linear Regression")
print("R2 Score:", round(multi_r2, 3))
```

```python
import joblib

# Save using joblib
joblib.dump(multi_model, 'multi_linear_joblib.pkl')
print("Multiple Linear Model saved using joblib")

```
